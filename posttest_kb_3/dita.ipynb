```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset_catagorical = pd.read_csv('supermarket_sales - Sheet1.csv')
```

```python
dataset_catagorical.head(10)
```

```python
x = dataset_catagorical['City']
y = dataset_catagorical['Unit price']
```

```python
plt.bar(x, y, color ='Yellow', width=0.5, bottom = 1)
plt.title('Sewa Di kota')
plt.xlabel('Kota')
plt.ylabel('Sewa')
plt.show()
```

```python
dataset_catagorical.corr()
```

```python
sns.heatmap(dataset_catagorical.corr(),cmap='Blues')
```

```python
labels = dataset_catagorical['City'].value_counts().rename_axis('kota').reset_index(name='City_Count')
```

```python
labels = pd.DataFrame(labels)
```

```python
labels.head(10)
```

```python
plt.pie(dataset_catagorical['City'].value_counts(), labels=labels['kota'], autopct='%.1f%%', startangle=90)
plt.show()
```
